# Model - Building

We are building the regression algorithm

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.metrics import plot_confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.model_selection import ShuffleSplit
from sklearn.metrics import r2_score
```

### Linear - Regression

 Note : linear regression is a regression model that estimates the relationship between one independent variable and one dependent variable using a straight line

```python
r2score_maxAccuracy=0
maxRandomState=0

for i in range(1, 100):
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=i)      # Creating train test split
    lr = LinearRegression()    # Creating Object of the Linear Regressor
    lr.fit(X_train, y_train)        # Fitting the x and y train data
    y_lr = lr.predict(X_test)       # predicting the x_test data and storing it into y_lr variable
    r2 = r2_score(y_test, y_lr)     #  r2_score is used to evaluate the performance of a Linear Regressor
        
    if r2>r2score_maxAccuracy:
        r2score_maxAccuracy=r2
        maxRandomState=i
        

print("------------------------------------------------")
print("                                                ")
print("R2 Score of Linear Regression is : ", r2score_maxAccuracy)
print("                                                ")
r2score_maxAccuracy_percentage = round(r2score_maxAccuracy,2)*100
print("R2 Score of Linear  Regression in Percentage is: ", r2score_maxAccuracy_percentage,"%")
print("                                                ")
if r2score_maxAccuracy_percentage < 70:
    print("R2 Score of Linear  Regression is Bad Model...")
elif r2score_maxAccuracy_percentage == 80:
    print("R2 Score of Linear  Regression is Model is Good...")
else:
    print("R2 Score of Linear  Regression is Best Model")
print("                                                ")
print("Random State", maxRandomState)
print("                                                ")
print("------------------------------------------------")
```

### Decision Tree Regressor

Note :  Decision tree regression observes features of an object and trains a model in the structure of a tree to predict data in the future to produce meaningful continuous output.

```python
r2score_maxAccuracy = 0
maxRandomState = 0

for i in range(1, 100):
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=i)    # Creating train test split
    dt = DecisionTreeRegressor()    # Creating Object of the Decision Tree Regressor
    dt.fit(X_train, y_train)        # Fitting the x and y train data
    y_dt = dt.predict(X_test)       # predicting the x_test data and storing it into y_dt variable
    r2 = r2_score(y_test, y_dt)     #  r2_score is used to evaluate the performance of a Decision Tree Regressor
    
    
    if r2>r2score_maxAccuracy:      
        r2score_maxAccuracy=r2
        maxRandomState=i
        

print("------------------------------------------------")
print("                                                ")
print("R2 Score of Decision Tree Regressor is : ", r2score_maxAccuracy)
print("                                                ")
r2score_maxAccuracy_percentage = round(r2score_maxAccuracy,2)*100
print("R2 Score of Decision Tree Regressor in Percentage is: ", r2score_maxAccuracy_percentage,"%")
print("                                                ")
if r2score_maxAccuracy_percentage < 70:
    print("R2 Score of Decision Tree Regressor is Bad Model...")
elif r2score_maxAccuracy_percentage == 80:
    print("Model is Good...")
else:
    print("Best Model")
print("                                                ")
print("Random State", maxRandomState)
print("                                                ")
print("------------------------------------------------")
```

### SVR (Support Vector Machine - Regressor)

Note : SVR uses the same basic idea as Support Vector Machine (SVM), a classification algorithm, but applies it to predict real values rather than a class. SVR acknowledges the presence of non-linearity in the data and provides a proficient prediction model.

```python
r2score_maxAccuray = 0
maxRandomState = 0

for i in range(1, 100):
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=i)      # Creating train test split
    svr = SVR()                         # Creating Object of the SVR (Support Vector Machine) Regressor
    svr.fit(X_train, y_train)           # Fitting the x and y train data
    y_svr = svr.predict(X_test)         # predicting the x_test data and storing it into y_svr variable
    r2 = r2_score(y_test, y_svr)        #  r2_score is used to evaluate the performance of a SVR (Support Vector Machine) Regressor
    
    if r2 > r2score_maxAccuracy:
        r2score_maxAccuracy = r2
        maxRandomState = i



print("------------------------------------------------")
print("                                                ")
print("Best R2 Score of SVR(Support Vector Machine) Regressor is : ", r2score_maxAccuracy)
print("                                                ")
r2score_maxAccuracy_percentage = round(r2score_maxAccuracy,2)*100
print("R2 Score of SVR(Support Vector Machine) Regressor in Percentage is: ", r2score_maxAccuracy_percentage,"%")
print("                                                ")
if r2score_maxAccuracy_percentage < 70:
    print("R2 Score of SVR(Support Vector Machine) Regressor is Bad Model...")
elif r2score_maxAccuracy_percentage == 80:
    print("Model is Good...")
else:
    print("Best Model")
print("                                                ")
print("Random State", maxRandomState)
print("                                                ")
print("------------------------------------------------")
```

### Random Forest Regression

Note : The Random Forest is a Regressor algorithm consisting of many decisions trees. It uses bagging and feature randomness when building each individual tree to try to create an uncorrelated forest of trees whose prediction by committee is more accurate than that of any individual tree

```python
r2score_maxAccuray = 0
maxRandomState = 0

for i in range(1, 100):
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=i)      # Creating train test split
    rf = RandomForestRegressor()                         # Creating Object of the RandomForest Regressor
    rf.fit(X_train, y_train)           # Fitting the x and y train data
    y_rf = rf.predict(X_test)         # predicting the x_test data and storing it into y_rf variable
    r2 = r2_score(y_test, y_rf)        #  r2_score is used to evaluate the performance of a RandomForest Regressor
    
    if r2 > r2score_maxAccuracy:
        r2score_maxAccuracy = r2
        maxRandomState = i



print("------------------------------------------------")
print("                                                ")
print("R2 Score of RandomForest Regressor is : ", r2score_maxAccuracy)
print("                                                ")
r2score_maxAccuracy_percentage = round(r2score_maxAccuracy,2)*100
print("R2 Score of RandomForest Regressor in Percentage is: ", r2score_maxAccuracy_percentage,"%")
print("                                                ")
if r2score_maxAccuracy_percentage < 70:
    print("R2 Score of RandomForest Regressor is Bad Model...")
elif r2score_maxAccuracy_percentage == 80:
    print("Model is Good...")
else:
    print("Best Model")
print("                                                ")
print("Random State", maxRandomState)
print("                                                ")
print("------------------------------------------------")
```
